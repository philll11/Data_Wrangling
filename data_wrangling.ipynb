# Data Wrangling

### Reading from a CSV file into a DataFrame data structure
The following script reads the data 'health_survey' file into a workable DataFrame object.

```python
import pandas as pd

%matplotlib inline

# Imports local csv files
wd = pd.read_excel("health_survey.xlsx", encoding="latin")
# Creates dataframe for easier workablility
healthSurveyData = pd.DataFrame(wd)

# Keeps columns between "population" and "percent.07"
healthSurveyData = healthSurveyData.loc[:, healthSurveyData.columns[0:11]]

# Changes column names
healthSurveyData.columns = ["population", "description", "group", "2016",
                           "2015", "2014", "2013", "2012", "2011", "2006", "2007"]

healthSurveyData.head()
```

Selecting 8 groups as features from the data for analysis.

```python
customData = healthSurveyData[((healthSurveyData["description"] == "Depression") 
                                                                      | (healthSurveyData["description"] == "Anxiety disorder")
                                                                      | (healthSurveyData["description"] == "Arthritis")
                                                                      | (healthSurveyData["description"] == "Cannabis use")
                                                                      | (healthSurveyData["description"] == "Definite confidence and trust in GP")
                                                                      | (healthSurveyData["description"] == "Current smokers"))
                                                                      & (healthSurveyData["group"] == "Total") 
                                                                      & (healthSurveyData["population"] == "adults")]


del customData["population"]
del customData["group"]

transposed = customData.transpose()
transposed.columns = transposed.loc["description"]
transposed.drop("description", inplace=True)
transposed.index.name = "Year"
tranSortedByYear = transposed.sort_index(ascending=True)

tranSortedByYear.head()
```

The following function takes as arguments, three different data filters, and a data frame and returns a new data frame containing the filtered data.

```python
def wrangle(_population, _group, _descriptions, _data):
    dfCollection = []
    for descrip in _descriptions:
        tempData = _data[(_data["description"] == descrip) & (_data["group"] == _group) & (_data["population"] == _population)]
        dfCollection.append(tempData)
    
    filteredData = pd.concat(dfCollection)
    del filteredData["population"]
    del filteredData["group"]

    transposedData = filteredData.transpose()
    transposedData.columns = transposedData.loc["description"]
    transposedData.drop("description", inplace=True)
    transposedData.index.name = "Year"
    tranSortedByYearData = transposedData.sort_index(ascending=True)
    
    return tranSortedByYearData
```

```python
descriptions1 = ["Depression", "Cannabis use", "Arthritis", "Current smokers", "Anxiety disorder"]
descriptions2 = ["Fizzy drink 1+ times per week", "Parent-rated health - excellent", "Dental health care worker visit"]
descriptions3 = ["Obese", "GP visit", "Hazardous drinkers (total population)", "Current smokers", "Gout", "Depression", "Cannabis use", "Arthritis"]
wrangle("adults", "Total", descriptions1, healthSurveyData)
wrangle("children", "Total", descriptions2, healthSurveyData)
wrangle("adults", "Total Pacific", descriptions3, healthSurveyData)
```

### Manual data wrangling

The following scripts reads baby names directly from the 'Smart Start' website and cleans it suitible for analysis.

```python
import pandas as pd
import io
import requests
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Retrieves the data
def RetrieveData(url):
    wd = pd.read_excel(url, sheet_name=[0,1], header=2)
    rawGDF = wd[0]
    rawBDF = wd[1]
    return rawGDF, rawBDF
```

```python
def YearsInDataSet(_df):
    # Retrieves the years for each name
    yearSerise = _df.loc[0][1:]
    yearList = []
    [yearList.append(x) for x in yearSerise if pd.notna(x)]
    return yearList
```

```python
# Cleans girls data
def Clean(_df, years):
    
    # NaN value features
    del _df["Unnamed: 0"]
    del _df["Unnamed: 1"]
    
    # There are 6 rows that contain no data that need to be removed
    _df.drop([0,1,2], inplace=True)
    _df.drop([105,104,103], inplace =True)
    
    # Variables
    numberHeading = "Unnamed: "
    columns = ["Name", "No.", "year", "sex"]
    namesDF = pd.DataFrame(columns=columns)
    j=0
    i=2
    while i <= len(_df.columns):
        columns = ["Name", "No.", "year", "sex"]
        tempDF = pd.DataFrame(columns=columns)
        tempDF["Name"] = _df[numberHeading + str(i)]
        i += 1
        tempDF["No."] = _df[numberHeading + str(i)]
        i += 2  
        tempDF["year"] = years[j]
        j += 1
        namesDF = pd.concat([namesDF, tempDF])
    
    return namesDF
```

```python
#Retrieve the data
url = "https://smartstart.services.govt.nz/assets/files/Top-100-girls-and-boys-names-since-1954.xlsx"
rawGDF, rawBDF = RetrieveData(url)

# Extract which years we are dealing with for each data set.
girl_year_list = YearsInDataSet(rawGDF)
boy_year_list = YearsInDataSet(rawBDF)
boy_year_list.remove(1985) # There are duplicate 1985 years

#Clean the data
girlsNamesDF = Clean(rawGDF, girl_year_list)
girlsNamesDF["sex"] = "Female"

boysNamesDF = Clean(rawBDF, boy_year_list)
boysNamesDF["sex"] = "Male"


# Joining dataframes
babyNamesDF = pd.concat([girlsNamesDF, boysNamesDF])
babyNamesDF

# Resetting index to 0
babyNamesDF.reset_index(inplace=True)
del babyNamesDF["index"]

# No. and year are not numeric so we must change them.
babyNamesDF[["No."]] = babyNamesDF[["No."]].apply(pd.to_numeric)


babyNamesDF.head()
# babyNamesDF.tail()
# babyNamesDF.year.unique()
# babyNamesDF.sex.unique()
# len(babyNamesDF)
```

```python
# Line graph
males = babyNamesDF[['Name', "No.", "year"]][babyNamesDF["sex"] == "Male"]
aarons = males[["No.", "year"]][males["Name"] == "Aaron"]
plot = aarons.plot(x="year", y="No.", title="Yearly totals for new borns named Aaron")
plot.set_ylabel("Totals")
```

The above graph displays the number of babies named Aaron over four decades.

A visual examination of this graph reveals a large spike in popularity for the name Aaron during the late 60's, early 70's. This is then followed by a sharp reduction in babies named Aaron.

The above graph compares the popularity of babies named Susan with babies named Ella.

A visual examination reveals that the popularity of both Susan and Ella has decreased dramatically over a 15-year period. We can see that Ella was the most popular name between the two expect for a single year in 1968.

#### Introduction

The following report will attempt to answer whether the affordability of house purchases for first home buyers is related to the amount of student loan debt a new graduate has.

This report will use the 'home_ownership_rate' and 'median_teriary_loan_balance_as_perc_of_med_inc_AHC' datasets. 
All datasets have be pulled from the 'nz_indicators.xlsx' dataset.

All extracted values have been cleaned and validated so that they are accurate and elegable for evaluation. Throughout this report I have included markdown code blocks that explains the cleaning process.

To answer this question effectively, I will be producing four plot that comparing different statistical measures within the two datasets.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
```

The 'nz_indicators' data set is full of errors that prevents us from effectively working with the data. Therefore it is important that we clean the data before we work with it.

The first cleaning method we will use is the 'na_values' parameter in the 'read_excel' method. This parameter takes an array of strings, finds any values in the data that match and replaces them with a NaN value.

```python
na_values = ["None", "NAN", "???", "-", "null", "NULL", "Null", "nan", "Nan"]
wd = pd.read_excel("nz_indicators.xlsx", na_values=na_values, encoding="latin")
# wd.head()
```

We need to clean the date and remove unwanted zeros at the end of each date.

We must also check the number of null values we have before and after we clean to validate our cleaning process.

```python
def formatDate(x):
    return x.replace(x[10:], '')

nullBefore = wd[["year"]].isna().sum()

wd['year'] = wd[['year']].applymap(formatDate)

nullAfter = wd[["year"]].isna().sum()
if not nullBefore.equals(nullAfter):
    print("Dataset does not match after cleaning")

# wd.head()
```

Next we want to set the date to be the index of the dataframe.

```python
wd.set_index('year', inplace=True)
wd.sort_index(inplace=True)
# wd.head()
```

Due to the nonuniformity of the values in the dataframe, we need to clean so that all valid values are floats.

```python
def removeChar(x):
    if pd.notna(x):
        if type(x) == float:
            return x
        elif type(x) == int:
            return float(x)
        else:
            return float(''.join(re.findall("[0-9]*[.]*[0-9]", x)))
    else:
        return x


nullBefore = wd.isna().sum()

wd = wd.applymap(removeChar)

nullAfter = wd.isna().sum()
if not nullBefore.equals(nullAfter):
    print("Dataset does not match after cleaning")

# wd.head()
```

Now that the data is cleaned, we need to extract the datasets we are going to be using.

```python
datasetNames = ["national_house_purchasing_HAM", "median_teriary_loan_balance_as_perc_of_med_inc_AHC"]

datasetLabels = ["Percentage of owner-occupied homes compared to total private dwelling", "Tertiary loan leaving balance as a percentage of income"]

df = wd[datasetNames]
df.columns = datasetLabels

df = df.dropna()
# df.head()
```

#### Data Analysis

```python
x1 = df["Percentage of owner-occupied homes compared to total private dwelling"]
x2 = df["Tertiary loan leaving balance as a percentage of income"]

plt.figure(figsize=(30,20))

plt.subplot(221)
plt.plot(x1)
plt.title("Percentage of owner-occupied homes compared to total private dwelling")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.xticks(rotation="60")

plt.subplot(222)
plt.plot(x2)
plt.title("Tertiary loan leaving balance as a percentage of income")
plt.xlabel("Year")
plt.ylabel("Percentage")
plt.xticks(rotation="60")
```

A visual examination of the above affordability of houses plot reveals a positive trend from 2003 till 2007, followed by a negative trend from 2007 till 2017.
A visual examination of the student loan balance plot reveals a negative trend from 2003 till 2010, followed by a sharp positive trend from 2010 till 2015.

It can be said that house affordability for first home buyers was increasing up until 2007 while student loan balances were decreasing up until 2010. During this three year period between 2007 and 2010, we see the trends reversing in both datasets.

There maybe undelying factors causing this reversal such as the 2008 finacial crisis. More rigorous analysis is needed to be certain.
